```python
from scipy.optimize import linprog 

# Coeficientes da função objetivo (minimizar) 

c = [-3, -5]  # queremos maximizar 3x + 5y → multiplicamos por -1 

# Restrições do tipo ≤ 

A_ub = [ 
    [2, 1], 
    [1, 2] 
] 

b_ub = [8, 8] 

# Limites das variáveis 
x_bounds = [(0, None), (0, None)] 

# Resolver 
res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=x_bounds, method='highs') 

print("Valor ótimo:", -res.fun)  # voltar o sinal 

print("Solução ótima:", res.x) 
```

```python
from scipy.optimize import linprog

# Coeficientes da função objetivo (minimizar)
c = [0.5, 0.2, 0.1]

# Restrições do tipo ≤
A_ub = [
    [-0.1, -0.2, -0.1],
    [-0.4, -0.4, -0.3],
    [0.2, 0.2, 0.1],
    [-0.2, 0, -0.4]
]

b_ub = [-0.12, -0.36, 0.36, 0]

# Restrições do tipo =
A_eq = [
    [0.1, 0.2, 0.1],
    [1, 1, 1]
]

b_eq = [0.18, 1]

# Limites das variáveis (x, y, z)
x_bounds = [(0, None), (0, None), (0, None)]

# Resolver
res = linprog(c,
              A_ub=A_ub, b_ub=b_ub,
              A_eq=A_eq, b_eq=b_eq,
              bounds=x_bounds,
              method='highs')

print("Valor ótimo:", res.fun)  # valor mínimo da função objetivo
print("Solução ótima:", res.x)  # valores de x, y, z
```

```python
from scipy.optimize import linprog

# Coeficientes da função objetivo (maximizar 1900x + 2100y)
# Como o linprog só minimiza, multiplicamos por -1
c = [-1900, -2100]

# Restrições do tipo ≤
A_ub = [
    [4, 2],
    [2, 3],
    [100, 200]
]

b_ub = [20, 10, 500]

# Limites das variáveis (x e y ≥ 0)
x_bounds = [(0, None), (0, None)]

# Resolver
res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=x_bounds, method='highs')

# Mostrar resultados
print("Valor ótimo:", -res.fun)   # inverter o sinal para refletir a maximização
print("Solução ótima:", res.x)    # valores de x e y
```
